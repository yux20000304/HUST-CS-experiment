## question 1

- 题目要求

假设我们使用的是sigmoid作为激活函数，需要我们写出梯度下降更新时$ w_{1,2} $的更新情况

- 思路

这里直接根据链式法则进行求解

 $$\frac{\partial loss}{\partial w_{1,2}^{[1]}}
        = \frac{\partial loss}{\partial o}\frac{\partial o}{\partial h_2}\frac{\partial h_2}{\partial w_{1,2}^{[1]}}$$
这里使用$g(x)$标记sidmoid函数
        $$g'(x) = g(x)(1-g(x))$$
因此就有$$\frac{\partial loss}{\partial w_{1,2}^{[1]}}
        =\frac{2}{m}\sum_{i=1}^m (o^{(i)}-y^{(i)})o^{(i)}(1-o^{(i)})w_2^{[2]}h_2^{(i)}(1-h_2^{(i)})x^{(i)}_1$$
其中$$h_2^{(i)}=g(x^{(i)}_1w_{1,2}^{[1]}+x^{(i)}_2w_{2,2}^{[1]}+w_{0,2}^{[1]})$$
        
        
b. 假设： $(0.5,0.5)$, $(3.5,0.5)$, $(0.5,3.5)$ 
神经网络的前向传输可以写成矩阵的形式：
        $$
        \left[\begin{matrix}
            -1.5 & 3 & 0 \\
            -1.5 & 0 & 3 \\
            9 & -3 & 3\\
        \end{matrix}\right] 
        \times
        \left[\begin{matrix}
            1 \\
            x_1 \\
            x_2\\
        \end{matrix}\right]
        $$
        有
        $$
        \left[\begin{matrix}
            -1 & -1 & -1 & 2.33
        \end{matrix}\right] 
        \times
        \left[\begin{matrix}
            1 \\
            h_1 \\
            h_2\\
            h_3\\
        \end{matrix}\right]
        $$
        如果一旦位于三角形中，则有：
        $$
        \left[\begin{matrix}
            1 \\
            1 \\
            1 \\
        \end{matrix}\right]
        $$ 
   所以第二个乘积是-0.67，最后的结果是0。

否则，第二个乘积将大于或等于0.33，最终结果为1。
