# 题目简介
本题需要我们应用监督学习技术去预测类星体的光谱。类星体是距离我们很遥远的发光体，这里我们将光强和光线的波长联系起来得到了下面的图，其中蓝色的代表固有光谱，红色的代表观测到的光谱.

这里可以很明显的看到观测信号相对于固有信号产生了衰减，在波长低于某一个值的时候光强会自动衰减一定的值，根据这个衰减模型建立出了以下的模型

$$ f_{obs}(\lambda) = absorption(\lambda)\cdot f(\lambda) + noise(\lambda) $$

### 问题a

使用局部加权线性回归

由题可知

$$ \sum_{i,j} W_{ij}(\Theta^Tx^{(i)}-y^{i})(\Theta^Tx^{(j)}-y^{j}) $$


最终需要证明的是
    
$$ W_{ii} = \frac{1}{2}w^{(i)} $$

首先
        $$\bigtriangledown_{\theta}J(\theta) = 2(W(X\theta-\vec y)^TX)$$
   让偏导数值为0
   $$ \theta =  (X^TWX)^{-1}X^TW\vec{y} $$
   $ \theta $即为我们所求的
   得到liklihood函数
    $$ l = \sum_{i=1}^m log\frac{1}{\sqrt{2\pi}\sigma^{(i)}} - \sum_{i=1}^m \frac{(y^{(i)}-\theta^Tx^{(i)})^2}{2(\sigma^{(i)})^2} $$
   这里有$$ w_{i}=\frac{1}{2(\sigma^{(i)})^2} $$ 

```python
#导入库
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df_train=pd.read_csv("./data/quasar_train.csv")
df_train.head()
```

```python
train_column=df_train.columns.values.astype(float).astype(int)
plt.plot(train_column,df_train.loc[0])
```

```python
df_train.shape
```

```python
df_test=pd.read_csv("./data/quasar_test.csv")
test_column=df_test.columns.values.astype(float).astype(int)
train_x=df_train.values.astype(float)
test_x=df_test.values.astype(float)
```

由之前的学习我们知道要让损失函数最小，即J对$\theta$求导值为0，从而得到了$\theta$的矩阵表达式
$$ \theta =(x^Tx)^{-1}x^{T}\vec{y} $$

```python
#问题1要求使用正态方程实现线性回归对模型1进行拟合
plt.title('answer 1')
x=np.stack([np.ones(train_x[0].shape),train_column]).T
y=train_x[0]
theta=np.linalg.inv(x.T.dot(x)).dot(x.T).dot(y)
plt.plot(train_column,train_x[0],label='sample')
plt.plot(train_column,train_column.dot(theta[1])+theta[0],label='regression')
plt.legend()
plt.show()
```

问题二加入了权重，因此$\theta$的值应该改为
$$ \theta =(x^Tx)^{-1}x^{T}W\vec{y} $$
这里w已经给出

```python
#问题二要求我们使用w权重来进行局部加权线性回归
plt.figure('answer 2')
regression=[]
tau=5
for t, x_evaluate in x:
    W=np.diag(np.exp(-(x_evaluate - x[:, 1]) ** 2 / (2 * tau ** 2)))
    theta = np.linalg.inv(x.T.dot(W).dot(x)).dot(x.T).dot(W).dot(y)
    regression.append(theta[1] * x_evaluate + theta[0])
plt.plot(train_column, train_x[0], label='sample')
plt.plot(train_column,regression,label='regression')
plt.legend()
plt.show()
```

问题三要求我们使用不同的$\tau$值进行输出

```python
x = np.stack([np.ones(train_x[0].shape), train_column]).T
y = train_x[0]
regression = []
taus = [1, 10, 100, 1000]
for i in range(len(taus)):
    regression.append(np.ones(len(train_column)))
for i in range(len(x)):
    x_evaluate= x[i][1]
    for j in range(len(taus)):
        tau = taus[j]
        W = np.diag(np.exp(-(x_evaluate - x[:, 1]) ** 2 / (2 * tau ** 2)))
        theta = np.linalg.inv(x.T.dot(W).dot(x)).dot(x.T).dot(W).dot(y)
        regression[j][i] = theta[1] * x_evaluate + theta[0]
fig, axes = plt.subplots(2, 2, figsize=(9, 6))
axes = axes.ravel()
for i in range(len(taus)):
    axes[i].set_title('tau = %d' % (taus[i]))
    axes[i].plot(train_column, train_x[0],label='sample')
    axes[i].plot(train_column, regression[i],label='regression')
    axes[i].legend()
plt.show()
```

任务四要求

题目大意：我们希望能够准确的预测行星的光强，但是由于前面的已知条件我们知道对于观测的光强，在波长小于1200和波长大于1300的情况是不同的，我们希望能够对观测到的右边的值对左边的光强分布情况进行预测，通过对含有噪声的右边的光强分布情况可以做出对左边光强情况的预测，这里给出了观测的光强公式
$$ f_{obs}(\lambda) = absorption(\lambda)\cdot f(\lambda) + noise(\lambda) $$
我们将函数回归任务作为学习函数r映射的目标
$$r(f_{right})(\lambda)=E(f_{left} | f_{right})(\lambda)$$
这里的第一步要求我们必须对训练数据集进行平滑处理这里直接使用上面我们加权线性回归的模型，$\tau$的值设置成5

```python
s_train = []
s_test = []
thetas = []
for i in range(train_x.shape[0]):
    regression = []
    tmp_thetas = []
    x = np.stack([np.ones(train_x[i].shape), train_column]).T
    y = train_x[i]
    tau = 5
    for t, x_evaluate in x:
        W = np.diag(np.exp(-(x_evaluate - x[:, 1]) ** 2 / (2 * tau ** 2)))
        theta = np.linalg.inv(x.T.dot(W).dot(x)).dot(x.T).dot(W).dot(y)
        regression.append(theta[1] * x_evaluate + theta[0])
        tmp_thetas.append(theta)
    thetas.append(tmp_thetas)
    s_train.append(regression)
for i in range(test_x.shape[0]):
    regression = []
    X = np.stack([np.ones(test_x[i].shape), test_column]).T
    Y = test_x[i]
    tau = 5
    for t, x_evaluate in X:
        W = np.diag(np.exp(-(x_evaluate - X[:, 1]) ** 2 / (2 * tau ** 2)))
        theta = np.linalg.inv(X.T.dot(W).dot(X)).dot(X.T).dot(W).dot(Y)
        regression.append(theta[1] * x_evaluate + theta[0])
    s_test.append(regression)
s_train = np.array(s_train)
s_test = np.array(s_test)
```

```python
plt.plot(train_column,s_test[0])
```

这里题目给出了一个算法，定义了以下的输入矩阵
$$ d(f_{1},f{2})=\sum_{i}(f_{1}(\lambda_{i})-f_{2}(\lambda_{i}))^{2} $$

```python
def ker(t):
    return max(1-t,0)
k = 3
estimation = []
errors = []
train_left, train_right = np.split(s_train,[150],axis=1)
for i in range(train_right.shape[0]):
    row = train_right[i]
    dist = ((train_right-row)**2).sum(axis=1)
    h = dist.max()

    neighbors = dist.argsort()[:k]

    bottom = 0
    for j in range(len(neighbors)):
        bottom += ker(dist[neighbors[j]]/h)
    left_hat = np.zeros(len(train_left[0]))
    for j in range(len(neighbors)):
        left_hat += ker(dist[neighbors[j]]/h)*train_left[neighbors[j]]
    left_hat /= bottom
    estimation.append(left_hat)
    error = np.sum((train_left[i] - left_hat) ** 2)
    errors.append(error)
errors = np.array(errors)
estimation = np.array(estimation)
print('The average error is %.6f'%errors.mean())


examples = [1,50,100,150]
fig, axes = plt.subplots(2,2,figsize=(10,6))
for i in range(len(examples)):
    idx = examples[i]
    ax = axes.ravel()[i]
    ax.plot(train_column, s_train[idx], label='smoothed')
    ax.plot(train_column,
            np.hstack([estimation[idx],
            np.zeros(max(0,np.abs(train_column.shape[0]-estimation[idx].shape[0])))]),
            label='estimation')
    ax.legend()
    ax.set_title('Training Set Sample #%d' % (idx))
plt.show()


estimation = []
errors = []
test_left, test_right = np.split(s_test,[150],axis=1)
for i in range(test_right.shape[0]):
    row = test_right[i]
    dist = ((train_right-row)**2).sum(axis=1)
    h = dist.max()

    neighbors = dist.argsort()[:k]

    bottom = 0
    for j in range(len(neighbors)):
        bottom += ker(dist[neighbors[j]]/h)
    left_hat = np.zeros(len(train_left[0]))
    for j in range(len(neighbors)):
        left_hat += ker(dist[neighbors[j]]/h)*train_left[neighbors[j]]
    left_hat /= bottom
    estimation.append(left_hat)
    error = np.sum((test_left[i] - left_hat) ** 2)
    errors.append(error)
errors = np.array(errors)
estimation = np.array(estimation)
print('The average error is %.6f'%errors.mean())


examples = [1,6,10,15]
fig, axes = plt.subplots(2,2,figsize=(10,6))
for i in range(len(examples)):
    idx = examples[i]
    ax = axes.ravel()[i]
    ax.plot(test_column, s_test[idx], label='smoothed')
    ax.plot(test_column,
            np.hstack([estimation[idx],
            np.zeros(max(0,np.abs(test_column.shape[0]-estimation[idx].shape[0])))]),
            label='estimation')
    ax.legend()
    ax.set_title('Test Set Sample #%d' % (idx))
plt.show()
```
