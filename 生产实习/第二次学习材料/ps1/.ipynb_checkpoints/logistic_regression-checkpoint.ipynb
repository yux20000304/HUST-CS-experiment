# Logistic Regression

**问题a**

题目中是需要我们证明对于任意向量z都有:

$$
\begin{align*}
z^T H z \geq 0
\end{align*}
$$

对于这个问题，需要我们首先写出hessian矩阵的表示形式，然后根据已经有的

$$
\begin{align*}
z^T H z = \sum_{i}^{}\sum_{j}^{} z_{i}H_{ij}z_{j}\geq 0 
\end{align*}
$$

最后完成证明，以下是证明过程

$$
\begin{align*}
    J(\theta)= \frac{1}{m} \sum_{i=1}^{m}log(1+e^{-y^{(i)}\theta^{T}x^{(i)}})
\end{align*}
$$

可以得到一阶导表达式

$$
\begin{align*}
    \frac{\partial J(\theta)}{\partial \theta_{i}}
            =  \frac{1}{m} \sum_{k=1}^{m} \frac{-y^{(k)}x^{(k)}_i}{1+e^{y^{(k)}\theta^{T}x^{(k)}}}
\end{align*}
$$

求出二阶导表达式
$$
\begin{align*}
    \frac{\partial J(\theta)}{\partial \theta_{i} \partial \theta_{j}}
            =  \frac{1}{m} \sum_{k=1}^{m} 
            \frac{x^{(k)}_ix^{(k)}_je^{y^{(k)}\theta^{T}x^{(k)}}}
            {(1+e^{y^{(k)}\theta^{T}x^{(k)}})^2}
\end{align*}
$$

把Hessian矩阵表达式H带入要证明的式子中

$$
\begin{align*}
    Z^THZ &= \sum_{i=1}^n \sum_{j=1}^n z_iH_{ij}z_j \\
            &= \frac{1}{m} \sum_{k=1}^{m} 
            \frac{\sum_{i=1}^n \sum_{j=1}^n z_ix^{(k)}_ix^{(k)}_jz_je^{y^{(k)}\theta^{T}x^{(k)}}}{(1+e^{y^{(k)}\theta^{T}x^{(k)}})^2}
\end{align*}
$$

通过对分子进行分析

$$
\begin{align*}
   \sum_{i=1}^n \sum_{j=1}^n z_ix^{(k)}_ix^{(k)}_jz_j = (X^TZ)^2 \geq 0 
\end{align*}
$$

$$
\begin{align*}
   \frac{e^{y^{(k)}\theta^{T}x^{(k)}}}{(1+e^{y^{(k)}\theta^{T}x^{(k)}})^2} > 0
\end{align*}
$$

又因为分母恒大于0，故表达式恒大于0，因此命题得证

**问题b**

给了我一个数据集，要求我使用牛顿法对逻辑回归进行收敛，理论上会比梯度下降法收敛的更快

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

#读取数据文件
df_x=pd.read_table("./data/logistic_x.txt",delim_whitespace=True, header=None)
df_y=pd.read_table("./data/logistic_y.txt",header=None)
df_y.astype(int)
x = df_x.values
y = df_y.values
m = len(x)
i=1
plt.title('Original Data')
for i in range(m):
    if y[i] == 1:
        plt.scatter(x[i][0], x[i][1], color='red')
    else:
        plt.scatter(x[i][0], x[i][1], color='blue')
plt.show()
theta = np.zeros(3)
```

```python
def fun1(y_k, theta, x_k):
    return 1.0 + math.exp(y_k * theta * x_k.T)
dfx = pd.read_table('data/logistic_x.txt',delim_whitespace=True, header=None)
dfy = pd.read_table('data/logistic_y.txt', header=None)
dfy.astype(int)
x = dfx.values
y = dfy.values
m = len(x)
plt.title('Original Data')
for i in range(m):
    if y[i] == 1:
        plt.scatter(x[i][0], x[i][1], color='red')
    else:
        plt.scatter(x[i][0], x[i][1], color='blue')
plt.show()

theta = np.zeros(3)

# working part
delta = 1e9
iter = 0
while delta > 1e-6:
    iter += 1
    # calculate the value
    der1 = np.mat(np.zeros(3))
    der2 = np.mat(np.zeros((3,3)))
    tmp_theta = np.mat(theta)

    for k in range(m):
        if y[k] == 1:   y_k = 1.0
        else:   y_k = -1.0
        x_k = np.mat(np.append(np.array(1),x[k]))
        bottom = fun1(y_k,tmp_theta,x_k)
        der1 += -y_k*x_k/bottom
        der2 += (math.exp(y_k * tmp_theta * x_k.T) / bottom**2)*np.mat(x_k.T*x_k)
    der1 /= m
    der2 /= m
    der2 = der2.I
    delta = np.abs(np.sum(der2*der1.T))
    tmp_theta = (tmp_theta.T - der2*der1.T).T
    theta = tmp_theta.getA()[0]

    # draw the line
    x1 = np.arange(0, 8, 0.1)
    plt.title('After %d Iterations' % (iter))
    for i in range(m):
        if y[i] == 1:
            plt.scatter(x[i][0], x[i][1], color='red')
        else:
            plt.scatter(x[i][0], x[i][1], color='blue')
    plt.plot(x1, -(theta[0] + theta[1]*x1) / theta[2] )
    plt.show()
    print('%d Iteration: theta changed by %.6f'%(iter,delta))
print('Converged After %d Iterations'%(iter))
print(theta)
```
